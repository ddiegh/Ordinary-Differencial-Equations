Considerando la ecuacion diferencial en una varibale $\frac{dx}{dt}$ = $\lambda x$ (un crecimiento proporcional) cuya solución es $x(t)$ = $x_0 e^{\lambda (t-t_0)}$ en donde $t_0$ es el tiempo inicial (normalmente no lo pondremos, es decir usamos $t_0 = 0$) y $x_0$ es la condicion inical (el valor de $x(t_0)$). Utilizando esta ecuacion diferencial podemos resolver ecuaciones en más dimensiones 

Consideremos la ecuacion diferencial que está definida por el campo vectorial $v(x_1, x_2) = (\lambda x_1, \lambda x_2)$ que es lo mismo que escribirla de la forma: 
$$
 \dot{\mathbf{x}} = A\mathbf{x} \implies 
\begin{bmatrix} \dot{x}_1 \\ \dot{x}_2 \end{bmatrix} = 
\begin{bmatrix} \lambda & 0 \\ 0 & \lambda \end{bmatrix} 
\begin{bmatrix} x_1 \\ x_2 \end{bmatrix} 
$$

Donde podemos observar que ninguna variable depende de la otra entonces podemos resolverlo de manera separada (como la ecuacion en $R$)
Es decir, las soluciones se ven de la forma 
$$
x(t) = (c_1 e^{\lambda_{1}t}, c_2 e^{\lambda_{2}t}) 
$$
en donde cada $c_i$ es la condicion inicial 

Esto se puede extrapolar a ecuaciones definidas de forma similar pero en más dimensiones 

Entonces para concluir: una ecuacion definida de la forma
$$
\begin{bmatrix} \frac{dx_{1}}{dt} \\ \frac{dx_{2}}{dt} \end{bmatrix} = 
\begin{bmatrix} \dot{x}_1 \\ \dot{x}_2 \end{bmatrix} = 
\begin{bmatrix} \lambda & 0 \\ 0 & \lambda \end{bmatrix} 
\begin{bmatrix} x_1 \\ x_2 \end{bmatrix} 
$$
tiene solución 
$$
x(t) = (c_1 e^{\lambda_{1}t}, c_2 e^{\lambda_{2}t}) 
$$

# Ejemplo: Sistema con Punto Silla 

Consideramos la ecuación diferencial vectorial:
$$\dot{\mathbf{x}} = \begin{bmatrix} 4 & 0 \\ 0 & -2 \end{bmatrix} \begin{bmatrix} x_1 \\ x_2 \end{bmatrix}$$

#### 1. Solución General
Como la matriz es diagonal, el sistema está desacoplado. La solución general es:
$$\varphi(t) = (c_1 e^{4t}, c_2 e^{-2t})$$

#### 2. Relación de Coordenadas
Para encontrar la trayectoria en el plano $x_1x_2$, eliminamos el parámetro $t$. 

Partiendo de $\varphi_1 = c_1 e^{4t}$:
$$\frac{\varphi_1}{c_1} = e^{4t} \implies \ln\left(\frac{\varphi_1}{c_1}\right) = 4t \implies t = \frac{1}{4} \ln\left(\frac{\varphi_1}{c_1}\right)$$

Sustituimos $t$ en la ecuación de $\varphi_2$:
$$\varphi_2 = c_2 e^{-2t}$$
$$\varphi_2 = c_2 e^{-2 \left[ \frac{1}{4} \ln\left(\frac{\varphi_1}{c_1}\right) \right]}$$
$$\varphi_2 = c_2 e^{-\frac{1}{2} \ln\left(\frac{\varphi_1}{c_1}\right)}$$
$$\varphi_2 = c_2 e^{\ln\left(\frac{c_1}{\varphi_1}\right)^{1/2}}$$

Simplificando, obtenemos la relación algebraica de las curvas:
$$\varphi2 = \frac{c_2 \sqrt{c_1}}{\sqrt{\varphi_1}} \implies \varphi_2 = \frac{k}{\sqrt{\varphi_1}}$$

#### 3. Análisis Asintótico
Para entender el comportamiento geométrico, analizamos los casos de las constantes:

* **Caso $c_1 = 0, c_2 \neq 0$:**
    $$\varphi(t) = (0, c_2 e^{-2t})$$
    La solución vive en el **eje $x_2$** y tiende al origen $(0,0)$ cuando $t \to \infty$.
    
* **Caso $c_2 = 0, c_1 \neq 0$:**
    $$\varphi(t) = (c_1 e^{4t}, 0)$$
    La solución vive en el **eje $x_1$** y se aleja del origen hacia el infinito cuando $t \to \infty$.

* **Caso $c_1 \neq 0, c_2 \neq 0$:**
    Las trayectorias son hipérbolas que se acercan a un eje y escapan por el otro.

#### 4. Comportamiento de la Norma
Calculamos la distancia de la solución al origen:
$$||\mathbf{x}(t)|| = \sqrt{c_1^2 e^{8t} + c_2^2 e^{-4t}}$$

Analizando los límites:
* $\lim_{t \to -\infty} ||\mathbf{\varphi}(t)|| = \infty$ (la solución viene del infinito por el término $e^{-4t}$)
* $\lim_{t \to \infty} ||\mathbf{\varphi}(t)|| = \infty$ (la solución se va al infinito por el término $e^{8t}$)

### 5. Análisis de Tangencia (Pendientes)
La pendiente de las curvas solucion está dada por:
$$\frac{\dot{\varphi_2}}{\dot {\varphi_1}} = \frac{-2c_2 e^{-2t}}{2c_1 e^{4t}} = \frac{- c_2}{2c_1 e^{6t}} $$

Entonces al analizar los limites
$$\lim_{t \to \infty}  \frac{\dot{\varphi_2}}{\dot{\varphi_1}} = 0$$
$$\lim_{t \to - \infty}  \frac{\dot {\varphi_2}}{\dot{\varphi_1}} = \infty $$
Esto explica visualmente por qué todas las trayectorias entran al origen de forma **vertical** (paralelas al eje $x_2$) y se alejan al infinito de con pendiente 0, es decir, paralelas al eje $x_1$

#### 6. Rapidez del Flujo (Magnitud del Campo)
La coloración de las trayectorias en la gráfica representa la rapidez del sistema. Lo que nos dice que tanto se está moviendo el sistema en distintos puntos del plano

* **El origen:** es un punto de equilibrio (velocidad nula).
* **Lejos del origen:** La rapidez aumenta linealmente con la distancia, lo que se observa en la gráfica con el cambio hacia colores más obscuros en los bordes.
**Conclusión:** El origen es un **Punto de Silla**, el cual es un equilibrio **inestable**.

## Un ejemplo en $R^3$ 

Considerando la ecuacion diferencial 
$$\dot{\mathbf{x}} = \begin{bmatrix} 2 & 0 & 0 \\ 0 & 2 & 0 \\ 0 & 0 & -2 \end{bmatrix} \begin{bmatrix} x_1 \\ x_2 \\  x_3 \end{bmatrix}$$
El procedimiento seria el mismo que en $R^2$ solo que el analisis se vuelve más complicado (raro?) pero llegamos a que la solucion a la ecuacion es: 
$$\phi(t) = (c_1 e^{2t}, c_2 e^{2t}, c_3 e^{-2t})$$

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')


fondo = '#1a1a2e'
fig.patch.set_facecolor(fondo)
ax.set_facecolor(fondo)
for axis in [ax.xaxis, ax.yaxis, ax.zaxis]:
    axis.set_pane_color((0.1, 0.1, 0.15, 1.0))
    axis.line.set_color('white')

theta = np.linspace(0, 2 * np.pi, 40) 
radio = 0.2

# Vector de tiempo (hacia adelante)
t = np.linspace(0, 1.6, 100) 


for th in theta:
    x0 = radio * np.cos(th)
    y0 = radio * np.sin(th)
    
    x_up = x0 * np.exp(2*t)
    y_up = y0 * np.exp(2*t)
    z_up = 4 * np.exp(-2*t)
    
    x_dn = x0 * np.exp(2*t)
    y_dn = y0 * np.exp(2*t)
    z_dn = -4 * np.exp(-2*t)

    ax.plot(x_up, y_up, z_up, color='#00d2ff', alpha=0.5, lw=1)
    ax.plot(x_dn, y_dn, z_dn, color='#ff007f', alpha=0.5, lw=1)

ax.plot([0, 0], [0, 0], [-5, 5], color='#ffffff', lw=3, label='$W^s$: Variedad Estable')
xx, yy = np.meshgrid(np.linspace(-5, 5, 2), np.linspace(-5, 5, 2))
zz = np.zeros_like(xx)
ax.plot_surface(xx, yy, zz, color='#fca311', alpha=0.15, edgecolor='none')

ax.set_title('Soluciones de la EDO', 
             color='white', pad=25, fontsize=15)
ax.set_xlabel('$x_1$', color='white', fontsize=12)
ax.set_ylabel('$x_2$', color='white', fontsize=12)
ax.set_zlabel('$x_3$', color='white', fontsize=12)
ax.tick_params(colors='white')

ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_zlim([-4.5, 4.5])
ax.legend(facecolor=fondo, edgecolor='white', labelcolor='white', loc='upper left')

plt.tight_layout()
plt.show()
```
